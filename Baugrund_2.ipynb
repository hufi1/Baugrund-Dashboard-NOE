```python
#import excel sheet which we created in the previous jupyter notebook
df1 = pd.read_excel(r'C:\Users\Sabrina\Desktop\Baugrundpreise\bodenpreise_noe.xlsx')
df2 = pd.read_excel(r'C:\Users\Sabrina\Desktop\Baugrundpreise\katgem.xlsx')
```

```python
#merging the 2 dfs on the unique "Gemeindenummer" and dropping the columns we don't need as well as the duplicates which are created in the process.
df3 = pd.merge(df1, df2, left_on='Gemeindenummer', right_on='GDNR', how='right')
df3 = df3.drop(['Katastralgemeinde', 'GDNR alt', 'GDNR', 'KGNR'], axis=1)
df3 = df3.drop_duplicates()
```

```python
#Check for success:
df3.head()
```

```python
#new column Ort_PLZ to uniquely identify the villages through the google API 
### (since a couple of the villages exist multiple times and we don't want them to be displayed in germany for example)
df3['Ort_PLZ'] = df3['PLZ_Ortsgemeinde'].astype('str') + ' ' + df3['Gemeindename']
```

```python
#Check for success:
df3.head()
```

```python
#set API key
gmaps_key = googlemaps.Client(key = 'YOUR API KEY')
```

```python
#creating 2 new columns for lat, lon. 


df3['Breitengrad'] = None
df3['Längengrad'] = None

#looping over each Ort_PLZ and passing it to the geocode API.
for i in range (0, len(df3), 1):
    geocode_result = gmaps_key.geocode(df3.iat[i, 10])
    try:
        #get these fields:
        lat = geocode_result[0]['geometry']['location']['lat']
        lon = geocode_result[0]['geometry']['location']['lng']
        #save them in the corresponding columns
        df3.iat[i, 11] = lat
        df3.iat[i, 12] = lon
    except:
        #in case it doesn't work:
        lat = None
        lon = None
```
