```python
import os
import pandas as pd
```

```python
cwd = os.path.abspath(r'C:\Users\Sabrina\Desktop\Baugrundpreise\bodenpreise je bezirk') # assign the path
```

```python
files = os.listdir(cwd)
```

```python
files #display the filenames within the folder
```

```python
#load the files in order to append them within one df
df = pd.DataFrame()
for file in files:
    if file.endswith('xlsx'):
        df = df.append(pd.read_excel('C:\\Users\\Sabrina\\Desktop\\Baugrundpreise\\bodenpreise je bezirk\\' + file, ignore_index = True))
```

```python
df['Grünlandpreis'] = df['Grünlandpreis'].replace('keine Angabe', 0)
df['Baulandpreis'] = df['Baulandpreis'].replace('keine Angabe', 0)
```

```python
df = df.set_index('Gemeindenummer')
```

```python
df = df.drop(index='Datenquelle: IMMOunited GmbH')
df = df.dropna()
```

```python
df['Einwohner']=df['Einwohner'].astype('str').replace('.000', '')
```
